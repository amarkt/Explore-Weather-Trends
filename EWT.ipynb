```python
import pandas as pd                 
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('results.csv')
```

```python
df.head()
```

```python
moving_avg = df.rolling(150,center=False,on='cat').mean().dropna()
```

```python
moving_avg
```

```python
plt.plot(moving_avg['year'],moving_avg['gat'],label='Global')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Temp')
plt.title("Global temp")
plt.show()
```

```python
plt.plot(moving_avg.year,moving_avg.cat,label='New Delhi')
plt.xlabel('Years')
plt.ylabel('Temp')
plt.title('New Delhi Temp')
plt.show()
```

```python
plt.plot(moving_avg.year,moving_avg.gat,label='Global')
plt.plot(moving_avg.year,moving_avg.cat,label='New Delhi')
plt.xlabel('Years')
plt.ylabel('Temp')
plt.title('Global vs New Delhi Temp over time')
plt.show()
```
